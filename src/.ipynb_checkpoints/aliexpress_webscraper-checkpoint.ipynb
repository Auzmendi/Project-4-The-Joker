ALIEXPRESS WEBSCRAPER

```python
import requests
from bs4 import BeautifulSoup

url = 'https://es.aliexpress.com/category/204011178/smart-electronics.html?category_redirect=1'

# Realizar la solicitud GET a la pgina
response = requests.get(url)

# Verificar si la solicitud fue exitosa (cdigo de estado 200)
if response.status_code == 200:
    # Obtener el contenido HTML de la respuesta
    html = response.content
    print(html)

    # Crear el objeto BeautifulSoup para analizar el HTML
    soup = BeautifulSoup(html, 'html.parser')
    print(soup.prettify())
    # Encontrar los elementos que contienen la informacin de los productos
    product_name = soup.find_all('div', class_='manhattan--content--1KpBbUi')
    print(len(product_name))
    product_name= str(product_name)
    print(product_name)
    # Recorrer cada elemento de producto y extraer la informacin deseada
    for product in product_name:
        nombre = product.find('h1', class_='manhattan--titleText--WccSjUS').text.strip()
        precio = product.find('div', class_='manhattan--price--WvaUgDY').text.strip()
        print(f'Nombre: {nombre}\nPrecio: {precio}\n')
else:
    print('No se pudo acceder a la pgina.')
```
